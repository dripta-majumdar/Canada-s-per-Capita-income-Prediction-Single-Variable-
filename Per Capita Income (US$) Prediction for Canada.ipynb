```python
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import linear_model
import numpy as np
import pandas as pd
```

```python
df= pd.read_csv("canada_per_capita_income.csv")
df.head()
```

```python
plt.xlabel("Year")
plt.ylabel("Per capita income(US$)")
plt.title("Yearly Per capita income of Canada")
plt.scatter(df.year, df.per_capita_income, marker="+")
# plt.plot(df.year, df.per_capita_income, color="blue")
```

```python
reg= linear_model.LinearRegression()
reg.fit(df[["year"]], df.per_capita_income)
```

```python
# Predicting the per capita income for the year 2020
reg.predict([[2020]])
```

```python
plt.xlabel("Year")
plt.ylabel("Per_capita_income(US$)")
plt.title("Per capita income of Canada")
plt.scatter(df.year, df.per_capita_income, marker="+", color="red")
plt.plot(df.year, reg.predict(df[["year"]]), color="blue")
```

```python
df2=pd.DataFrame({
    "year":[2020, 2021, 2022, 2023, 2024, 2025, 2026, 2027, 2028, 2029, 2030]
})
df2
```

```python
predictions=reg.predict(df2[["year"]])
predictions
```

```python
df2["Per_capita_income_predictions(US$)"]= predictions
df2.drop("Per_income_capita_predictions(US$)", axis="columns")
```

```python
df2.to_csv("2020-2030 Per Capita Income (US$) Prediction for Canada.csv", index=False)
```
